# Từ trang web gốc lấy url 20 đội bóng

```python
url = 'https://fbref.com/en/comps/9/2023-2024/2023-2024-Premier-League-Stats'
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')
table = soup.find('table', {'id': 'results2023-202491_overall'})
team_names = []
team_links = []
rows = table.find('tbody').find_all('tr')

for row in rows:
    team_cell = row.find('td', {'data-stat': 'team'})
    if team_cell:
        team_name = team_cell.text.strip()
        team_names.append(team_name)
        link_tag = team_cell.find('a')
        if link_tag and 'href' in link_tag.attrs:
            team_link = 'https://fbref.com' + link_tag['href']
        else:
            team_link = None
        team_links.append(team_link)

df = pd.DataFrame({
    'team': team_names,
    'url': team_links
})

df.to_csv('squad_url.csv', index=False)
```

```python
df
```

### chia thành 2 dataframe để dễ dàng crawl data

```python

num_rows = len(df)
half = num_rows // 2
df1 = df.iloc[:half]
df2 = df.iloc[half:]
```

### Nửa số đội bóng đầu

```python
teams_df = df1
all_data = []
for index, row in teams_df.iterrows():
    team_name = row['team']
    url = row['url']
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    table = soup.find('table', {'id': lambda x: x and x.startswith('stats_standard')})
    if table is not None:
        df = pd.read_html(str(table))[0]
        df['team'] = team_name
        all_data.append(df)
    else:
        print(f"Could not find the stats table for {team_name}")
combined_df = pd.concat(all_data, ignore_index=True)
combined_df.head()
```
